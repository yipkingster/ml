**Jax prohibits non-array (list, tuples, etc) input to APIs, unlike numpy**

```python
from jax import make_jaxpr
import jax.numpy as jnp

# Not allowed, unlike numpy
# print(jnp.sum([1,2,3]))

# Allow conversion
arr = jnp.array([1,2,3])
print(jnp.sum(arr))

#See jaxpr
def permissble_sum(x):
  return jnp.sum(jnp.array(x))

print(permissble_sum([1,2,3]))
make_jaxpr(permissble_sum)([1,2,3])

```
