**Jax pseudorandom number usage**

```python
from jax import random
import jax.numpy as jnp

key = random.key(43)
print(key)

print(random.normal(key))
# Same random number generated.
print(random.normal(key))

for i in range(3):
  # The key must be splited before passing the normal()
  newkey, subkey = random.split(key)
  print(random.normal(subkey))
  del subkey

  key = newkey
```
